# Clustering with UMAP + DBSCAN Starter Notebook

```python
import numpy as np 
import pandas as pd 
import umap 
from sklearn.cluster import DBSCAN 
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt 
import seaborn as sns 
```

# Step 2: Load and prepare data

```python
data = load_iris()
X = data.data  # features
y_true = data.target  # true labels (for reference only, clustering is unsupervised)
```

```python
print("Data shape:", X.shape)
```

# Step 3: Reduce dimensions with UMAP (from 4D to 2D for easy visualization)

```python
reducer = umap.UMAP(n_neighbors=15, min_dist=0.1, random_state=42)
X_umap = reducer.fit_transform(X)
```

<!-- Breakdown of parameters in UMAP(...):
n_neighbors=15
This controls how many neighboring points UMAP looks at when trying to learn the local structure of the data.
Smaller values focus on local details (small-scale structure).
Larger values consider broader/global structure.
min_dist=0.1
This controls how tightly UMAP packs points together in the low-dimensional space.
Smaller min_dist means points can be closer, preserving more local clusters.
Larger min_dist spreads points out more, emphasizing global structure.
random_state=42
This sets the random seed so results are reproducible. Using the same seed means the output wonâ€™t change every time you run it.

What about fit_transform(X)?
This tells UMAP to learn the mapping from your high-dimensional data X to a lower dimension (default is 2D), and apply that transformation immediately, returning the transformed data as X_umap.

X_umap will be a 2D array with the same number of rows as X but with only 2 columns (or however many dimensions you set), representing the data in a simpler space that preserves structure.
 -->


```python
print("UMAP shape:", X_umap.shape)
```

# Step 4: Apply DBSCAN clustering on UMAP output
# eps and min_samples are DBSCAN parameters you can tune

```python
dbscan = DBSCAN(eps=0.5, min_samples=5)
clusters = dbscan.fit_predict(X_umap)

print("Cluster labels:", clusters)
```

# Step 5: Visualize clusters

```python
plt.figure(figsize=(8,6))
palette = sns.color_palette('bright', np.unique(clusters).max() + 1)
colors = [palette[x] if x >= 0 else (0.5, 0.5, 0.5) for x in clusters]  # gray for noise
```

```python
plt.scatter(X_umap[:, 0], X_umap[:, 1], c=colors, s=50, alpha=0.8)
plt.title('UMAP + DBSCAN Clustering of Iris Data')
plt.xlabel('UMAP dimension 1')
plt.ylabel('UMAP dimension 2')
plt.show()
```

# Compare clusters to true labels (just to see how well clustering did)

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=X_umap[:, 0], y=X_umap[:, 1], hue=y_true, palette='deep', s=50)
plt.title('UMAP Projection Colored by True Labels')
plt.show()
```
